## **2.0 Introduction: The 14-Day Race**
> *"In public health, time is the only currency that matters. A 24-hour delay in response can result in an exponential surge in infections; a 14-day head start can save a city."*

Cholera remains a persistent shadow over Kenya’s public health landscape, particularly within the Lake Victoria Basin and urban informal settlements. Despite significant advances in medical treatment, the **National Multi-sectoral Cholera Elimination Plan (2022-2030)** identifies a critical failure: our surveillance systems are reactive. We detect outbreaks only after hospitals are overwhelmed.

**K-CEWS (Kenya Cholera Early Warning System)** flips this script. By utilizing **spatiotemporal machine learning**, we bridge the gap between "Climate Shocks" (heavy rainfall and flooding) and the biological transmission cycle of *Vibrio cholerae*. K-CEWS provides a predictive "Sentinel" that looks 14 days into the future, allowing the Ministry of Health to move from managing crises to preventing them.


## **3.0 Problem Statement**
The current epidemiological surveillance in Kenya relies on the Integrated Disease Surveillance and Response (IDSR) framework. While robust for tracking, it lacks predictive foresight.

* **The Lag Gap:** There is a documented correlation between hydro-climatic triggers (rainfall) and cholera surges, yet this data is not integrated into real-time health alerts.
* **Resource Inefficiency:** Emergency response (vaccine deployment and WASH kits) is currently triggered by "Case Spikes," which means resources often arrive after the peak of the transmission curve.
* **Data Silos:** Environmental data (NASA), health data (WHO), and demographic data (KNBS) exist in isolation, preventing a holistic understanding of outbreak risk.


## **4.0 Aims and Objectives**
### **4.1 General Objective**
To develop an AI-driven predictive surveillance system (**K-CEWS**) that utilizes hydro-climatic and socio-demographic data to forecast cholera outbreak probabilities 14 days in advance at the sub-county level in Kenya.

### **4.2 Specific Objectives**
* To integrate and preprocess diverse data streams from NASA POWER, WHO Global Health Observatory, and KNBS Census reports.
* To implement **Feature Lagging (t-14 days)** to model the relationship between climate triggers and clinical case onset.
* To develop a multi-model predictive pipeline benchmarking **Negative Binomial Regression**, **Random Forest**, and **XGBoost**.
* To deploy a high-availability **Streamlit Dashboard** for real-time risk visualization and decision support for SCHMTs.

## **6.0 Justification & Significance**
The implementation of K-CEWS offers a three-fold benefit:
* **Life-Saving Foresight:** Provides a "Window of Opportunity" for pre-positioning Oral Cholera Vaccines (OCV).
* **Economic Efficiency:** Proactive prevention is significantly cheaper than emergency hospitalizations and large-scale disinfection campaigns.
* **Data-Driven Policy:** Empowers the Ministry of Health with an evidence-based tool to justify resource allocation to high-risk "Hotspots."


## **7.0 Scope and Limitations**
* **Scope:** The project focuses on high-risk reservoirs including the Lake Victoria Basin (Kisumu, Migori, Homa Bay) and Nairobi’s informal settlements. Data spans 2018 to April 2026.
* **Limitations:** Accuracy is dependent on the consistency of reporting in the KHIS (DHIS2) system. Unexpected human migrations (refugee movements) may introduce variables outside of climatic triggers.


## **8.0 Definition of Key Terms**
* **Spatiotemporal:** Data that encompasses both geographic location (Space) and time-stamped intervals (Time).
* **Feature Lagging (t-14):** The process of shifting independent variables (rainfall) back by 14 days to align with the future dependent variable (cholera cases).
* **Overdispersion:** A statistical property where the variance is higher than the mean, common in epidemic data where cases "spike" suddenly.


## **9.0 Ethical Considerations & Data Governance**
The development of K-CEWS adheres to the **Data Protection Act (2019) of Kenya** and international ethical standards.

### **9.1 De-identification and Anonymization**
* **The K-CEWS Protocol:** We operate strictly at the Sub-County/Aggregate level. The model does not ingest or store "Line Lists" containing names or exact household GPS coordinates. By using aggregated incidence rates (cases per 100k), we protect individual privacy.

### **9.2 Avoiding Community Stigmatization**
* **The K-CEWS Protocol:** Our dashboard is designed as a Decision-Support Tool for Health Officials, not a public-facing map. Access is restricted to authorized SCHMT members to ensure data is used for intervention rather than isolation.

### **9.3 Algorithmic Bias & Equity**
* **The K-CEWS Protocol:** We utilize Negative Binomial Regression to handle sparse data and incorporate KNBS socio-demographic baselines to ensure low-reporting rural areas are not ignored by the model.


## **10.0 Project Evaluation & Success Metrics**

To ensure the reliability of K-CEWS as a decision-support tool, the project will be evaluated based on statistical rigor, geographic focus, and model transparency.

### **10.1 Statistical Performance Metrics**
We will utilize standard regression metrics to quantify the deviation between the predicted cholera incidence and the actual reported cases:
* **Mean Absolute Error (MAE):** To understand the average "error gap" in our case count predictions.
* **Root Mean Squared Error (RMSE):** To penalize large errors, ensuring the model does not significantly underestimate "Superspreader" events or massive surges.

### **10.2 Milestone Definition (MVP vs. Stretch Goal)**
The project is structured in two tiers of success:
* **Minimum Viable Product (MVP):** A 7-day predictive model successfully deployed for the high-risk reservoirs of the Lake Victoria Basin.
* **Stretch Goal (Explainable AI):** Integration of **SHAP (Lundberg & Lee)** values. This transitions the system from a "Black Box" to a transparent tool, allowing the dashboard to explain *why* a sub-county is flagged (e.g., *"Risk is high due to 48-hour rainfall exceeding the 90th percentile"*).

### **10.3 Qualitative Success Matrix**
Beyond the math, the project is considered successful if it achieves:
1. **User Utility:** A health officer can interpret the Risk Dial in under 10 seconds.
2. **Latency:** The model can ingest new NASA data and update the forecast within a 5-minute window.

### **11.2 Data Extraction Logic Applied**
To ensure the "14-Day Race" (Section 2.0) is technically possible, the following pre-processing steps were applied during acquisition:

* **IMERG Satellite Integration:** We transitioned from standard precipitation to **NASA GPM-IMERG** data for higher spatial resolution in the Lake Victoria Basin.
* **Metadata Cleaning:** NASA headers were stripped, and `-999` null values were flagged for mean-imputation to ensure model stability.
* **Risk Factor Digitization:** Socio-demographic data was digitized from the *2022-2030 National Cholera Elimination Plan* research, turning qualitative "High/Medium" risk levels into quantitative weights for the XGBoost model.
* **Coordinate Precision:** Coordinates for Kisumu (-0.09, 34.77) and Migori (-1.06, 34.47) were snapped to the MERRA-2 grid for environmental consistency.

> **Lead's Note:** The "Testing Set" (actual_cases_2026_test.csv) has been isolated from the training data to ensure an unbiased evaluation of the 14-day predictive accuracy.

## **Chapter 2: Literature Review**

### **2.1 Global Benchmark: The NASA-UNICEF Yemen Project**
* **The Project:** A high-level collaboration using satellite data to predict cholera in Yemen.
* **Successes:** Achieved 92% accuracy using NASA environmental "signatures."
* **Gaps:** Faced a "War-Data Gap"—unreliable ground reporting prevented real-time verification.
* **K-CEWS Capitalization:** Kenya’s stable KHIS/DHIS2 reporting allows for superior model verification compared to the Yemen study.

### **2.2 Local Benchmark: The Kenya Red Cross sEAP Protocol**
* **The Project:** The Simplified Early Action Protocol (sEAP) for Cholera in Kenya.
* **Successes:** Established clear environmental triggers (e.g., rainfall >50mm/day).
* **Gaps:** Relies on manual checks of rain gauges, making it "Human-Heavy" and slow.
* **K-CEWS Capitalization:** We are automating these triggers using Machine Learning for instantaneous, data-driven alerts.

### **2.3 Technical Benchmark: Nigeria Machine Learning Study (2026)**
* **The Project:** "Optimizing Cholera Outbreak Prediction in Nigeria."
* **Successes:** Proved XGBoost explained 69% of case variance.
* **Gaps:** Lacked "Lead Time"—models predicted the *current* week rather than providing a future window.
* **K-CEWS Capitalization:** By adding the **14-day Feature Lag**, we provide the "Foresight Window" that previous studies lacked.

### **2.4 Summary of Gaps & K-CEWS Innovation**

| Limitation in Previous Work | K-CEWS Solution (Our Capitalization) |
| :--- | :--- |
| **Reactive Monitoring:** Models only show current status. | **Predictive Lag:** Implementing a t-14 day lag for early warning. |
| **Environmental Isolation:** Models ignore population density. | **Holistic Integration:** Combining NASA (Rain) + KNBS (Sanitation). |
| **Lack of User Interface:** Models exist only in code. | **Streamlit Command Center:** Visual dashboard for health officers. |
